# Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

```python
df = pd.read_csv("student-mat.csv", sep=";")
```

# Analysis Questions

```python
average_grade = df["G3"].mean()

print("Average Final Grade:", average_grade)
```

```python
high_score = df[df["G3"] > 15]

print("Students scoring above 15:", len(high_score))
```

```python
correlation = df["studytime"].corr(df["G3"])

print("Correlation:", correlation)
```

```python
df.groupby("sex")["G3"].mean()
```

# Data Visualization

# histogram

```python
plt.figure(figsize=(8,5))

plt.hist(df["G3"], bins=10, edgecolor="black")

plt.title("Distribution of Final Grades")

plt.xlabel("Final Grade")

plt.ylabel("Number of Students")

plt.savefig("grade_distribution.png")

plt.show()
```

# Scatter Plot

```python
plt.figure(figsize=(8,5))

plt.scatter(df["studytime"], df["G3"])

plt.title("Study Time vs Final Grade")

plt.xlabel("Study Time")

plt.ylabel("Final Grade")

plt.savefig("studytime_vs_grade.png")

plt.show()
```

# bar chart

```python
plt.figure(figsize=(6,4))

sns.barplot(x="sex", y="G3", data=df)

plt.title("Average Final Grade by Gender")

plt.savefig("gender_performance.png")

plt.show()
```

# correlation heatmap

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.savefig("correlation_heatmap.png")

plt.show()
```
